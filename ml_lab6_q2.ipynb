```python
df.plot.scatter(x = 'Weight', y = 'Size', s = 100)
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
x = df.iloc[:,0:2] # 1t for rows and second for columns
x
```

```python
kmeans = KMeans(3)
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
data_with_clusters = df.copy()
data_with_clusters['Clusters'] = identified_clusters 
plt.scatter(data_with_clusters['Weight'],data_with_clusters['Weight'],c=data_with_clusters['Clusters'],cmap='rainbow')
```

```python
data_with_clusters = df.copy()
data_with_clusters['Clusters'] = identified_clusters 
plt.scatter(data_with_clusters['Size'],data_with_clusters['Size'],c=data_with_clusters['Clusters'],cmap='rainbow')
```
