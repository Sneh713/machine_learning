# Transposing Arrays and Swapping Axes

```python
arr9=np.arange(15).reshape((3,5))
arr9
```

```python
arr9.T
```

```python
arr10=np.random.randn(6,3)
arr10
```

```python
arr10.T
```

```python
np.dot(arr10.T,arr10)
```

```python
arr11=np.arange(16).reshape((2,2,4))
arr11
```

```python
arr11.transpose((1,0,2))
```

```python
arr11.swapaxes(1,2)
```

```python
arr11.swapaxes(1,0)
```
